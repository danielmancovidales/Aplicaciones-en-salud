```python
## Cambiar la letra de mayuscula a miniscula.
df_egresos.columns=df_egresos.columns.str.lower()
df_egresos = df_egresos.applymap(lambda x: x.lower() if isinstance(x, str) else x)
df_egresos.head(2)
```

```python
# Se convierten las variables de fechas a formato datetime
df_egresos['fecha_nacimiento'] = pd.to_datetime(df_egresos['fecha_nacimiento']).dt.strftime('%Y-%m-%d')
df_egresos['fecha_nacimiento'] = pd.to_datetime(df_egresos['fecha_nacimiento'])
df_egresos['fecha_ingreso_clinica'] = pd.to_datetime(df_egresos['fecha_ingreso_clinica'])
df_egresos['fecha_ingreso_servicio'] = pd.to_datetime(df_egresos['fecha_ingreso_servicio'])
df_egresos['fecha_alta_medica'] = pd.to_datetime(df_egresos['fecha_alta_medica'])
df_egresos['fecha_salida'] = pd.to_datetime(df_egresos['fecha_salida'])
```

```python
# Se eliminan los nulos de la fecha de ingreso a la clinica
df_egresos= df_egresos.dropna(subset=['fecha_ingreso_clinica'])
```

```python
# Se toman las variables de fecha ingreso a la clinica y fecha de salida para procesarla como la variable dependiente
df_egresos[['fecha_ingreso_clinica','fecha_salida']]
```

```python
# Se calcula el tiempo de estancia de los pacientes en el hospital
df_egresos['tiempo_estancia']=(df_egresos['fecha_salida']-df_egresos['fecha_ingreso_clinica'] ).dt.total_seconds() / 3600
df_egresos['tiempo_estancia']
```

```python
# se deciden eliminar las fechas ya que fecha_ingreso_clinica y fecha_salida por que se se les hizo un procesamiento para conocer la estancia de los pacientes en horas, por lo cual estas variables no eran necesarias
# Se eliminaron el resto de fechas ya que no se consideran importantes teniendo en cuenta la información que se tiene de los pacientes
df_egresos=df_egresos.drop([ 'fecha_ingreso_servicio', 'fecha_alta_medica',
       'fecha_salida', 'fecha_nacimiento'], axis=1)
```
