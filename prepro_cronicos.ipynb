```python
#Eliminación de variables con nombres parecidos
patron = re.compile("^diagnostico.*")
```

```python
# Excluir una columna específica por nombre
columna_a_mantener = ['diagnostico_principal', 'diagnostico_epoc']
```

```python
# Filtrar y eliminar columnas que coincidan con el patrón, excepto la columna a mantener
columnas_a_eliminar = [col for col in df_cronicos.columns if patron.match(col) and col != columna_a_mantener]
```

```python
# Eliminar las columnas
df_cronicos = df_cronicos.drop(columns=columnas_a_eliminar)
```
